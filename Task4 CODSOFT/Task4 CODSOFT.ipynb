**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**Load Dataset**

```python
df = pd.read_csv('/content/drive/MyDrive/Projects/CODSOFT/advertising.csv')
```

**Data PreProcessing**

```python
df.isnull().sum()
```

```python
# Explore the data
print(df.head())
print(df.describe())
```

**Vizualization**

```python
sns.pairplot(df, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', kind='scatter')
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot histogram for 'TV'
df['TV'].plot.hist(bins=10, color='blue', ax=axes[0])
axes[0].set_title('TV')
axes[0].set_xlabel('TV')
axes[0].set_ylabel('Frequency')

# Plot histogram for 'Radio'
df['Radio'].plot.hist(bins=10, color='green', ax=axes[1])
axes[1].set_title('Radio')
axes[1].set_xlabel('Radio')
axes[1].set_ylabel('Frequency')

# Plot histogram for 'Newspaper'
df['Newspaper'].plot.hist(bins=10, color='orange', ax=axes[2])
axes[2].set_title('Newspaper')
axes[2].set_xlabel('Newspaper')
axes[2].set_ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

**Split Data**

```python
# Split the data into train and test sets
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

**Create a Model**

```python
# Create and fit the model
model = LinearRegression()
model.fit(X_train, y_train)
```

**Model Evaluation**

```python
# Evaluate the model
print('R-squared:', model.score(X_test, y_test))
```

**Prediction**

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred
```

```python
# Plot the predictions
plt.scatter(y_test, y_pred)
plt.xlabel('True Sales')
plt.ylabel('Predicted Sales')
plt.show()
```

**Custom testing**

```python
x_input = [50, 30, 10]

# Make a prediction
y_pred = model.predict([x_input])

# Print the prediction
print('Predicted Sales:', y_pred)
```
