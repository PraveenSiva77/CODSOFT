**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Load Dataset**

```python
# Load the credit card fraud dataset
df = pd.read_csv('/content/drive/MyDrive/Projects/CODSOFT/creditcard.csv')
df.head()
```

**Data PreProcessing**

```python
df.isnull().sum()
```

```python
print(df.describe())
```

**Vizualization**

```python
# Visualize the data
sns.pairplot(df, x_vars=['V1','V2','V3','V4','V5'], y_vars=['Amount'])
plt.show()
sns.pairplot(df, x_vars=['V6','V7','V8','V9','V10'], y_vars=['Amount'])
plt.show()
sns.pairplot(df, x_vars=['V24','V25','V26','V27','V28'], y_vars=['Amount'])
plt.show()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

**Split Data**

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

**Create a Model**

```python
# Train the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

**Model Evaluation**

```python
# Evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
